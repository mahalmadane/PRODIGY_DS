```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
adult=pd.read_csv("adult.csv")
adult.head(10)
```

```python
a=adult[["age","gender"]]
a.head(100)
```

```python
# verifier s'il ya pas de valeur maquant
print(a[['gender', 'age']].isnull().sum())
```

```python
AG=a[0:10320]
AG
```

```python
# Filtrer les valeurs manquantes
AG_clean = AG.dropna(subset=['gender', 'age'])

# Tracer le graphique
plt.bar(AG_clean["gender"], AG_clean["age"])
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Distribution de l'âge par genre")
plt.show()
```

```python
sns.displot(x="age",data=AG_clean,kde=True)
plt.show()
```
