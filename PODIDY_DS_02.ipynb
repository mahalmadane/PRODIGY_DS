```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
adult=pd.read_csv("data/adult.csv")
adult
```

### Data type

```python
adult.dtypes
```

### Histogramme

```python
plt.figure(figsize=(15,10))
#Age histogramme
plt.subplot(2,3,1)
sns.histplot(x='age',data=adult,kde=True)
# Eduction num
plt.subplot(2,3,2)
sns.histplot(x='educational-num',data=adult,kde=True)
# Capital gain
plt.subplot(2,3,3)
sns.histplot(x='capital-gain',data=adult,kde=True)
# Capital loss
plt.subplot(2,3,4)
sns.histplot(x='capital-loss',data=adult,kde=True)
plt.subplots_adjust(wspace=0.7, hspace=1)
```
